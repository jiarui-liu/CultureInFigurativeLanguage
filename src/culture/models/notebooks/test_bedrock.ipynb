```python
# Copyright Amazon.com, Inc. or its affiliates. All Rights Reserved.
# SPDX-License-Identifier: Apache-2.0
"""
Shows how to generate a message with Anthropic Claude using Converse API.
"""
import boto3
import logging
import os
import json


# Initialize Bedrock Runtime
bedrock = boto3.client("bedrock-runtime", region_name="us-east-1")

# Replace with your inference profile ID or ARN
# inference_profile_id = "global.anthropic.claude-sonnet-4-5-20250929-v1:0"
inference_profile_id = "us.anthropic.claude-sonnet-4-5-20250929-v1:0"

body = json.dumps({
    "messages": [
        {"role": "user", "content": "Hello from Python, can you respond?"}
    ],
    "max_tokens": 200,
    "anthropic_version": "bedrock-2023-05-31",
    # other parameters like max_tokens if needed
})

response = bedrock.invoke_model(
    modelId=inference_profile_id,
    body=body,
    contentType="application/json",
    accept="application/json"
)

result = json.loads(response["body"].read())['content'][0]['text']
print(result)
```
