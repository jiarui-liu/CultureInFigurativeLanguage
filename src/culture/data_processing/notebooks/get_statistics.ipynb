```python
import pandas as pd

# path to your txt file
file_path = culture_root + "/culture/data/chengyu_data/chengyu_definition.txt"

# read the file line by line
rows = []
with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        # split only on the first comma
        idiom, sentence = line.split(",", 1)
        rows.append({
            "idiom": idiom,
            "sentence": sentence
        })

# create DataFrame
df = pd.DataFrame(rows)

df
```
